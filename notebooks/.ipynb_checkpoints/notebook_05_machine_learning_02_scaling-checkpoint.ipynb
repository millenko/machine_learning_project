# train_test_split

```python
%run "common_imports.py"

df = pd.read_csv("../data/02.csv")
features = df.drop(columns=["is_canceled"])
target = df["is_canceled"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)

display(X_train.head(), "")
display(y_train.head(), "")
```

# scaler

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_train_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

X_test_scaled = scaler.transform(X_test)
X_test_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

display(X_train_df.head(), "")
display(X_test_df.head(), "")
```

## KNeighborsClassifier
- unscaled 79.04%, **scaled 81.79%**

```python
knn = KNeighborsClassifier(n_neighbors=2)
display(knn.fit(X_train_scaled, y_train))

pred = knn.predict(X_test_scaled)
print(f"The accuracy of the model is {knn.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred = pred, y_true = y_test), "\n")
```

## LogisticRegression
- **unscaled: max_iter=4000 (3000 wasn't enough), score 80.84%**
- top 5 features by absolute coefficient
  - required_car_parking_spaces 6.26
  - deposit_type_Non_Refund 3.46
  - previous_cancellations 2.83
  - deposit_type_No_Deposit 2.15
  - deposit_type_Refundable 1.52
  - 
- scaled, max_iter=4000 (even 1000 is enough, yields same score), score 80.14%
- top 5 features by absolute coefficient
  - previous_cancellations 24.14
  - required_car_parking_spaces 22.02
  - avg_daily_rate 8.47
  - previous_bookings_not_canceled 6.32
  - adults 5.07

```python
log_reg = LogisticRegression(max_iter=4000)
display(log_reg.fit(X_train_scaled, y_train), "")

pred = log_reg.predict(X_test_scaled)

print(f"Logistic Regression score is {log_reg.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred, y_true=y_test), "\n")

# individual coefficients
coefficients = log_reg.coef_[0]
feature_names = X_train.columns
coefficients_df = pd.DataFrame({"Feature": feature_names, "Coefficient": coefficients})
coefficients_df["Absolute Coefficient"] = abs(coefficients_df["Coefficient"])
coefficients_df[["Coefficient", "Absolute Coefficient"]] = coefficients_df[["Coefficient", "Absolute Coefficient"]].round(2)
coefficients_df = coefficients_df.sort_values(by="Absolute Coefficient", ascending=False)
display(coefficients_df)
```

## DecisionTreeClassifier
- **unscaled and scaled, same results**
- **score 81,93%**
- **tree importance**:
  - 'deposit_type_Non_Refund': '0.50',
  - 'lead_time': '0.10',
  - 'total_of_special_requests': '0.10',

  - 'arrival_date_week_number': '0.01',
  - 'adults': '0.01',
  - 'previous_bookings_not_canceled': '0.01',

```python
tree = DecisionTreeClassifier(max_depth=10)
display(tree.fit(X_train_scaled, y_train), "")

pred = tree.predict(X_test_scaled)

print(f"Decision Tree accuracy is {tree.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred, y_true=y_test), "\n")

tree_importance = {feature: f"{importance:.2f}" for feature, importance in zip(X_train.columns, tree.feature_importances_)}
sorted_tree_importance = {k: v for k, v in sorted(tree_importance.items(), key=lambda item: item[1], reverse=True)}
display(sorted_tree_importance, "")

tree = DecisionTreeClassifier(max_depth=2)
tree.fit(X_train_scaled, y_train)

dot_data = export_graphviz(tree, out_file="tree.dot", filled=True, rounded=True, feature_names=X_train_scaled.columns)

with open("tree.dot") as f:
    dot_graph = f.read()
graphviz.Source(dot_graph)
```

## Ensemble

### BaggingClassifier
- n_estimators=100, max_samples = 1000, what's the accuracy?
- KNeighborsClassifier(n_neighbors=2), unscaled 70.80%, **scaled** 78.73%
- LogisticRegression(max_iter=4000), **unscaled** 80.90%, scaled 77.87%
- **THE BEST SCORE: DecisionTreeClassifier(max_depth=10), unscaled 81.74%**, scaled 81.65%
- RandomForestClassifier, unscaled 81.40%, **scaled** 81.43%
- GradientBoostingClassifier, **unscaled** 81.31%, scaled 81.29%

```python
bagging_knn = BaggingClassifier(KNeighborsClassifier(n_neighbors=2), n_estimators=100, max_samples=1000)
display(bagging_knn.fit(X_train_scaled, y_train), "")
pred = bagging_knn.predict(X_test_scaled)

print(f"Bagging KNN accuracy is {bagging_knn.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred, y_true=y_test), "\n")
```

```python
bagging_log_reg = BaggingClassifier(LogisticRegression(max_iter=4000), n_estimators=100, max_samples=1000, verbose=3)
display(bagging_log_reg.fit(X_train_scaled, y_train), "")
pred = bagging_log_reg.predict(X_test_scaled)

print(f"Bagging LR accuracy is {bagging_log_reg.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred, y_true=y_test), "\n")
```

```python
bagging_dt = BaggingClassifier(DecisionTreeClassifier(max_depth=10), n_estimators=100, max_samples = 1000)
display(bagging_dt.fit(X_train_scaled, y_train), "")
pred = bagging_dt.predict(X_test_scaled)

print(f"Bagging DT accuracy is {bagging_dt.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred, y_true=y_test), "\n")
```

```python
bagging_rf = BaggingClassifier(RandomForestClassifier(), n_estimators=100, max_samples=1000, verbose=3)
display(bagging_rf.fit(X_train_scaled, y_train), "")
pred_rf = bagging_rf.predict(X_test_scaled)

print(f"Bagging RF accuracy is {bagging_rf.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred_rf, y_true=y_test), "\n")
```

```python
bagging_gb = BaggingClassifier(GradientBoostingClassifier(), n_estimators=100, max_samples=1000, verbose=3)
display(bagging_gb.fit(X_train_scaled, y_train), "")
pred_gb = bagging_gb.predict(X_test_scaled)

print(f"Bagging Gradient Boosting accuracy is {bagging_gb.score(X_test_scaled, y_test) * 100:.2f}%\n")
print(classification_report(y_pred=pred_gb, y_true=y_test), "\n")
```
