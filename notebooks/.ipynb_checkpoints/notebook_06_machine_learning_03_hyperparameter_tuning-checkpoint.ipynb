```python
%run "common_imports.py"

df = pd.read_csv("../data/02.csv")
features = df.drop(columns=["is_canceled"])
target = df["is_canceled"]

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.20, random_state=0)

display(X_train.head(), "")
display(y_train.head(), "")
```

# scaler

```python
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_train_df = pd.DataFrame(X_train_scaled, columns=X_train.columns)

X_test_scaled = scaler.transform(X_test)
X_test_df = pd.DataFrame(X_test_scaled, columns=X_test.columns)

display(X_train_df.head(), "")
display(X_test_df.head(), "")
```

# Hyperparameter tuning

## GridSearchCV

### KNeighborsClassifier
- 1st: **unscaled, best params: "algorithm": "auto", "n_neighbors": 9, "weights": "distance"; accuracy 80.32%, ct 8min**
- scaled, best params: "algorithm": "kd_tree", "n_neighbors": 9, "weights": "distance"; **accuracy 83.37%, ct 30min**
- observed: algorithm "ball_tree" takes very long time calculating scaled data vs unscaled

- 2nd: ct supposedly 6m, accuracy 81.13%,
- best params: "algorithm": "auto", "leaf_size": 10, "n_neighbors": 9, "p": 1, "weights": "distance"
-
- 3rd: ct: start 6.10 , accuracy
- best params:

```python
# 1st

grid = {"n_neighbors": range(1,10),
        "weights": ["uniform", "distance"],
        "algorithm": ["auto", "ball_tree", "kd_tree", "brute"]}

model = GridSearchCV(KNeighborsClassifier(), param_grid=grid, cv=5, scoring="accuracy", verbose=3, n_jobs=-1)

display(model.fit(X_train, y_train), "")

display("Best parameters:", model.best_params_ , "")

accuracy = model.best_estimator_.score(X_test, y_test) * 100
print(f"The accuracy of the tuned model is {accuracy:.2f}%")

pred = model.best_estimator_.predict(X_test)
print(classification_report(y_true=y_test, y_pred=pred))
```

```python
# 1st 2114

grid = {"n_neighbors": range(1,10),
        "weights": ["uniform", "distance"],
        "algorithm": ["auto", "ball_tree", "kd_tree", "brute"]}

# Start timing the execution of the GridSearchCV
%time model = GridSearchCV(KNeighborsClassifier(), param_grid=grid, cv=5, scoring="accuracy", verbose=3, n_jobs=-1)

# Fit the model and display the result
%time display(model.fit(X_train_scaled, y_train), "")

# Display the best parameters
%time display("Best parameters:", model.best_params_ , "")

# Predict using the best estimator and display the accuracy
%time pred = model.best_estimator_.predict(X_test_scaled)
%time print(f"The accuracy of the tuned model is {model.score(X_test_scaled, y_test)*100: .2f}%")

# Display classification report
%time print(classification_report(y_pred=pred, y_true=y_test))
```

```python
# 2nd

knn_params = {"n_neighbors": range(1, 10),
              "weights": ["uniform", "distance"],
              "algorithm": ["auto", "ball_tree", "kd_tree", "brute"],
              "leaf_size": [10, 20, 30],
              "p": [1, 2]}

knn_grid_search = GridSearchCV(KNeighborsClassifier(),
                               param_grid=knn_params,
                               cv=5,
                               scoring="accuracy",
                               verbose=3,
                               n_jobs=-1)

knn_grid_search.fit(X_train, y_train)

display("Best parameters:", knn_grid_search.best_params_)

pred = knn_grid_search.best_estimator_.predict(X_test)

accuracy = knn_grid_search.score(X_test, y_test) * 100
print(f"The accuracy of the tuned model is {accuracy:.2f}%")
print(classification_report(y_true=y_test, y_pred=pred))
```

```python
# 3rd

knn_params = {"n_neighbors": range(1, 10),
              "weights": ["uniform", "distance"],
              "algorithm": ["auto", "ball_tree", "kd_tree", "brute"],
              "leaf_size": [10, 20, 30],
              "p": [1, 2]}

knn_grid_search = GridSearchCV(KNeighborsClassifier(),
                               param_grid=knn_params,
                               cv=5,
                               scoring="accuracy",
                               verbose=3,
                               n_jobs=-1)

knn_grid_search.fit(X_train_scaled, y_train)

display("Best parameters:", knn_grid_search.best_params_)

pred = knn_grid_search.best_estimator_.predict(X_test_scaled)

accuracy = knn_grid_search.score(X_test_scaled, y_test) * 100
print(f"The accuracy of the tuned model is {accuracy:.2f}%")
print(classification_report(y_true=y_test, y_pred=pred))
```

### LogisticRegression
- best parameters: {"C": 10, "max_iter": 1000, "penalty": "l2", "random_state": 42, "solver": "sag"}
- accuracy 80.82%
- computing time 2m

```python
param_grid_outer = {"solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"],
                    "max_iter": [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000],
                    "random_state": [42]}

grid_search_outer = GridSearchCV(LogisticRegression(),
                                 param_grid_outer,
                                 cv=5,
                                 scoring="accuracy",
                                 verbose=3,
                                 n_jobs=-1)

grid_search_outer.fit(X_train_scaled, y_train)

best_params_outer = grid_search_outer.best_params_
best_solver = best_params_outer["solver"]
best_max_iter = best_params_outer["max_iter"]

param_grid_inner = {
    "C": [0.001, 0.01, 0.1, 1, 10],
    "penalty": ["l1", "l2"],
    "solver": [best_solver],
    "max_iter": [best_max_iter],
    "random_state": [42]
}

grid_search_inner = GridSearchCV(LogisticRegression(),
                                 param_grid_inner,
                                 cv=5,
                                 scoring="accuracy",
                                 verbose=3,
                                 n_jobs=-1)

grid_search_inner.fit(X_train_scaled, y_train)

best_params_inner = grid_search_inner.best_params_

print("Best parameters from outer grid search:", best_params_outer)
print("Best parameters from inner grid search:", best_params_inner)

final_model = LogisticRegression(**best_params_inner)
final_model.fit(X_train_scaled, y_train)

final_model_accuracy = final_model.score(X_test_scaled, y_test) * 100
print(f"Accuracy of the final model: {final_model_accuracy:.2f}%")
print(classification_report(y_pred=pred, y_true=y_test))
```
